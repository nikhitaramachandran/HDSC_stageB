```python
import numpy as np
import pandas as pd
```

```python
# Load the dataset 
df = pd.read_csv("C:/Users/nikhi/Downloads/energydata_complete.csv")
df
```

```python
df.info()
```

```python
# Question 12
X = np.array(df['T2']).reshape(-1, 1) # since we have a single feature
y = df['T6']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_squared = r2_score(y_test,pred)
```

```python
print(round(r_squared,2))
```

```python
# Question 13-17 
new_df = df.drop(columns = ['date', 'lights'])
```

```python
new_df
```

```python
col = new_df.columns
col
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
normalized_df = pd.DataFrame(scaler.fit_transform(new_df), columns=col)
```

```python
# Splitting the data into x and y
X = normalized_df.drop('Appliances',axis =1)
y = normalized_df['Appliances']
```

```python
# splitting data into train and test 

X_train, X_test, y_train, y_test =train_test_split(X,y,test_size =0.3, random_state =42)
```

```python
# Since already declared previously model = LinearRegression, using it directly
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, pred)
round(mae, 2) 
```

```python
rss = np.sum(np.square(y_test - pred))
round(rss, 2) 
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, pred))
round(rmse, 3) 
```

```python
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, pred)
round(r2_score, 2) 
```

```python
# Question 18
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(X_train, y_train)
pred = ridge_reg.predict(X_test)
```
